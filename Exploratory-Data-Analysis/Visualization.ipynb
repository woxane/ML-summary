# Section 0: Visualization
Visualization is a good techniuqe to get basic idea about your data.

In this notebook we use seaborn and matplotlib in python.

let's first import our packages and load the data set.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

file_path = "../data/iris_data.csv"
data = pd.read_csv(file_path)
```

```python
data.head()
```

Let's take a look at their histograms plot for a feature too see it's distribution of each feature

```python
# data['sepal_length'].plot.hist()
sns.histplot(data['sepal_length'], kde=True)
```

we can show the bar plot of the features to compare their mean values them using `sns.barplot(data)`.

```python
# sns.barplot(data)
sns.barplot(data['species'])
```

using scatter plots we can explore the relationship or correlation between two numerical variables.

```python
# sns.scatterplot(data)
sns.scatterplot(data, x='sepal_length', y='sepal_width')
# data.plot.scatter(x='sepal_length', y='sepal_width')
```

boxplot is useful for identifying outliers and comparing distributions across different categories.

```python
sns.boxplot(data)
# plt.boxplot(data['sepal_width'])
# data.plot.box()
# data.boxplot()
# data['sepal_width'].plot.box()
```

heatmaps are often used for showing the data correlations.

```python
sns.heatmap(data.drop(['species'], axis=1).corr())
```

joinplot Combines scatter plots and histograms (or KDE plots) to give a comprehensive view of bivariate data.

```python
sns.jointplot(data)
```

`sns.pairplot(data)` helps us to understand how variables are related to each other and the relationship between dependent and independent variables present in the dataset.

```python
sns.pairplot(data)
```
