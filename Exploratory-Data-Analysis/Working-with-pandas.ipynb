In this notebook we will start using pandas library in python and learn how to work with them.

### Reading the data
we will start using used card dataset for start.

```python
import pandas as pd

file_path = "../data/used_cars_data.csv"
data = pd.read_csv(file_path)
```

for a quick look at the datas we use `.head()` for first 5 observations or `.tail()` for last 5 observations method.

```python
# data.tail()
data.head()
```

we can use `.info()` to understand the data type and information about data.

from the below informations we could that that New_Price Column have to much null values.

```python
# We could use the above line if we want to just find the number of null for each columns.
# data.isnull().sum()
data.info()
```

to find unique values in each column we can use `.nunique()` method.

**This method helps us in feature engineering as we see in future.**

```python
data.nunique()
```

suppose you have useless columns and they don't help for our analysis you can delete them using `.drop()` method.

axis=1 for columns and axis=0 for indexs.

```python
data = data.drop(['S.No.'], axis = 1)
data.info()
```

You can convert your continues values to some groups like bins using `pd.cut()`.

```python
data['Km'] = pd.cut(data['Kilometers_Driven'], 10)
data['Yr'] = pd.cut(data['Year'], 5)
data
```
